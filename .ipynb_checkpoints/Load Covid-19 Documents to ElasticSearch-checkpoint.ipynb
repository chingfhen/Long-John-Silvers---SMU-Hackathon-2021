# ABOUT:
- Key steps:
    1. read json covid 19 document
    2. split document into 100 length passages
        - summaries are not split
    3. compute embedding for each passage
    4. load into elastic search

```python
import os
import utils
import preprocess
import encoders
import datastore
import yaml
```

```python
with open(r"C:\Users\tanch\Documents\Coding Competitions\SMU\SMU HACKATHON 2021\config.yaml") as f:
    config = yaml.load(f, Loader=yaml.FullLoader)
    print(config)
```

```python
datastore.datastore.delete_all_documents(config['index'])
os.chdir(config['directory'])
for filename in os.listdir():
    x = utils.read_document(filename)
    x = preprocess.add_datetime(x)
    x = preprocess.split_document(x, 150)
    x = encoders.embed_documents(x)
    datastore.datastore.write_documents(x,config['index'])
```

```python
docs = datastore.datastore.get_all_documents(config['index'])
```
